## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

# Cataboost

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print('All libraries imported')

#read the data
url = 'https://raw.githubusercontent.com/npradaschnor/Pima-Indians-Diabetes-Dataset/refs/heads/master/diabetes.csv'
data = pd.read_csv(url)
print(data.head(5))

print(data.info())
# no missing values but we have 0 values in some columns

# check the number of 0 values in each column
print((data==0).sum())

col_with_0 = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# replace with median
for col in col_with_0:
    data[col] = data[col].replace(0, data[col].median())

# check the number of 0 values in each column
print('\nAfter replacing 0 with median:')
print((data==0).sum())
print(data.head(5))
print(data.info())

# the distribution of outcome variable
print(data['Outcome'].value_counts())

# check the correlation
data['Outcome'].value_counts().plot(kind='bar')
# plt.show()

# check the correlation
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True)
# plt.show()

# check the correaltion for diagnosis for all the features or columns using corr
# Create a DataFrame
df = pd.DataFrame(data)
# Choose the feature to compare (e.g., 'feature1')
cont_feature = 'Outcome'
# Compute correlation of all features with the chosen feature
correlations = df.corr()[cont_feature]
print(correlations.sort_values())


# here we can feature selection and drop columns which have less correlation with diagnosis
# create independent and dependent variable
X=data.drop(['Outcome'],axis=1)
y=data['Outcome']

#create train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

# check the shape of the data
print('Shape of X_train:', X_train.shape)
print('Shape of X_test:', X_test.shape)

# apply statndard scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
# catboost
from catboost import CatBoostClassifier

# create object of GradientBoostingClassifier
cb = CatBoostClassifier(iterations=100, learning_rate=0.1, depth=2, random_state=0)
cb.fit(X_train, y_train)

print('Training completed')

# predict the test and train accuracy
y_pred_cb = cb.predict(X_test)
y_pred_train_cb = cb.predict(X_train)

# Get train and test score
from sklearn.metrics import confusion_matrix, accuracy_score
print('Confusion matrix: \n', confusion_matrix(y_test, y_pred_cb))
print('Accuracy score test: ', accuracy_score(y_test, y_pred_cb))
print('Accuracy score train: ', accuracy_score(y_train, y_pred_train_cb))
```
